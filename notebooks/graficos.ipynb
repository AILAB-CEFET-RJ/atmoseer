```python
import matplotlib.pyplot as plt, numpy as np
```

```python
# Dados
estacoes = ['A602', 'A621', 'A627', 'A636', 'A652']
# Valores do coeficiente para cada run de inmet, inmet + tpw, inmet + glm, inmet + glm + tpw
inmet = [0.5619442112353035, 0.49793209746597794, 0.4765120953055447, 0.6150110049522262, 0.5812602689594473]
inmet_tpw = [0.641844602076215, 0.5048708816057309, 0.5647846543612016, 0.5725260055520719, 0.5676682200268022]
inmet_glm = [0, 0.5776734670943755, 0.6210646596664263, 0.5873401903189216, 0.5556579840867885]  # Replaced None with 0
inmet_glm_tpw = [0, 0.5841945698700843, 0.6378493357134936, 0.5976183629203811, 0.5444521053389343]  # Replaced None with 0

barWidth = 0.15  # Largura das barras

# Configurando as posições das barras
r1 = np.arange(len(inmet))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# Configurando o tamanho do gráfico
plt.figure(figsize=(10, 6))

# Criando as barras
plt.bar(r1, inmet, color='#ff6361', width=barWidth, edgecolor='grey', label='INMET')
plt.bar(r2, inmet_tpw, color='#bc5090', width=barWidth, edgecolor='grey', label='INMET+TPW')
plt.bar(r3, inmet_glm, color='#58508d', width=barWidth, edgecolor='grey', label='INMET+GLM')
plt.bar(r4, inmet_glm_tpw, color='#003f5c', width=barWidth, edgecolor='grey', label='INMET+GLM+TPW')

# Adicionando títulos e rótulos
plt.title('Comparação do Coeficiente de Matthews por Estação', fontweight='bold')
plt.xlabel('Estações', fontweight='bold')
plt.ylabel('Coeficiente de Matthews', fontweight='bold')

# Adicionando nomes das estações no eixo X
plt.xticks([r + barWidth for r in range(len(r1))], estacoes)

# Adicionando legenda e grade
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.grid(False)

# Exibindo o gráfico
plt.show()
```
